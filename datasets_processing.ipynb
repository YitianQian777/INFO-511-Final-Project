### Identify the missing values for each dataset and then discuss how to deal with them accordingly.

```python
# for ev_population_data
# subset the data to contain rows that have at least one NA value
ev_population_data_NA = ev_population_data[ev_population_data.isnull().any(axis = 1)]
ev_population_data_NA.head(10)
```

```python
# for ev_population_size_history_county
# subset the data to contain rows that have at least one NA value
ev_population_size_history_county_NA = ev_population_size_history_county[ev_population_size_history_county.isnull().any(axis = 1)]
ev_population_size_history_county_NA.head(10)
```

```python
# for ev_population_size_history
# subset the data to contain rows that have at least one NA value
ev_population_size_history_NA = ev_population_size_history[ev_population_size_history.isnull().any(axis = 1)]
ev_population_size_history_NA.head(10)
```

```python
# for ev_WA_tax_exemptions
# subset the data to contain rows that have at least one NA value
ev_WA_tax_exemptions_NA = ev_WA_tax_exemptions[ev_WA_tax_exemptions.isnull().any(axis = 1)]
ev_WA_tax_exemptions_NA.head(10)
```

There are no NA rows in the datasets ev_population_size_history and ev_WA_tax_exemptions. So let's examine closely to the first two datasets: ev_population_data and ev_population_size_history_county.

```python
print(ev_population_data_NA.info())
```

The column that has the most missing values is "Legislative District", having 474 missing values. Other columns have 3 to 18 missing values. After looking through the ev_population_data_NA subset, we realized that for the rows that have missing values for "Legislative District", the values for the column "Electric Utility" are mostly "NON WASHINGTON STATE ELECTRIC UTILITY"; while for the rows that have valid values for "Legislative District", the values for the column "Electric Utility" have various values other than "NON WASHINGTON STATE ELECTRIC UTILITY", for instance, "BONNEVILLE POWER ADMINISTRATION||PUD NO 2 OF PACIFIC COUNTY", "PUGET SOUND ENERGY INC||CITY OF TACOMA - (WA)", etc. This makes sense because it might be difficult to track the legislative district if the electric utility is not from Washington state, since this dataset is collected from Washington.  

Since such connection exists, it would be information loss if we simply remove these rows with NA values for "Legislative District". We would keep these rows but replace the NA values with a string "Missing" to indicate that there are no valid values from the original dataset. For other columns, however, the missing values have no visible connection or meaning. Since those are potentially individual mistakes while collecting the data, we decide to just remove these rows.

To conclude, for the dataset ev_population_data, we would replace missing values for the variable "Legislative District" with the string value "Missing". Then we would remove all rows that have any missing value for other variables.

```python
print(ev_population_size_history_county_NA.info())
```

The columns that have missing values are only "County" and "State". All other columns have no missing values. After examining these rows, there are no visible connections among other columns, except for the fact that all these rows have the value of "Passenger" for the column "Vehicle Primary Use". However, this is not a significant connection, because in the original dataset, there are only two distinct values for this column: "Passenger" and "Truck". And there are 84% of the rows have the values of "Passenger". So even though we can say that all "Truck" rows have valid data for "County" and "State", it is not a very significant observation that contributes to any of our research questions.

Therefore, we would simply remove the rows that have missing values from the original dataset ev_population_size_history_county.

### Deal with the datasets accordingly based on the decision above.

```python
# deal with NA values for ev_population_data
# replace missing values for the variable "Legislative District" with the string value "Missing"
ev_population_data["Legislative District"] = ev_population_data["Legislative District"].fillna("Missing")

# remove all rows that have any missing value for other variables
ev_population_data = ev_population_data.dropna()

print(ev_population_data.info())
```

```python
# deal with NA values for ev_population_size_history_county
# remove the rows that have missing values
ev_population_size_history_county = ev_population_size_history_county.dropna()

print(ev_population_size_history_county.info())
```
